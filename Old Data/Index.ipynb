```python
common=set(DFtxt2023.columns)&set(DFtxt2022)&set(DFtxt2021)&set(DFtxt2020)

# we are keeping just the 'common' columns:
DFtxt2023ap=DFtxt2023.loc[:,common]
DFtxt2022ap=DFtxt2022.loc[:,common]
DFtxt2021ap=DFtxt2021.loc[:,common]
DFtxt2020ap=DFtxt2020.loc[:,common]
```

```python
theZips=zip(DFtxt2023ap.dtypes,
            DFtxt2022ap.dtypes,
            DFtxt2021ap.dtypes,
            DFtxt2020ap.dtypes)
# you see:
list(theZips)
```
